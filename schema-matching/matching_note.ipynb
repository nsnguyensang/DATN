```python
alo_min = alo[["description","price","link_image","date","type"]]
bat_min = batdongsanso[["content","price","link_image","date","type"]]
ibat_min = ibat[["brief","price","link_image","date","type"]]
nha24_min = nha24[["description","price","link_image","date","type"]]
```

```python
alo_min = alo_min.sample(n=1000)
bat_min = bat_min.sample(n=700)
ibat_min = ibat_min.sample(n=1000)
nha24_min = nha24_min.sample(n=1000)
```

```python
from valentine import valentine_match, valentine_metrics
from valentine.algorithms import JaccardLevenMatcher
matcher = JaccardLevenMatcher(0.2)


matches = valentine_match(ibat_min, nha24_min , matcher)

print(matches)
```
